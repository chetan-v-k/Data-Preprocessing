Data Analysis - Automatic Categorical

```python
# Why Automate Categorical Analysis?
# When analyzing categorical data, such as app-categories or content ratings, we often need to count the number of occurrences of each unique category. Doing this manually for every category is not practical.

# For example:

# Finding the number of apps in a category (e.g., "Art & Design") can be done manually, but what if we need the count for all 33 categories?
# Instead of writing the same logic multiple times, we can automate the process using loops and dictionaries.
```

```python
#Toal number of apps in each Category
#Doing the Categorical Analysis with all the columns(categories)
print(len(df['Category'].unique()))

categories = {}
for name in df['Category'].unique():
    c=0
    for i in df['Category']:
        if(i==name):                            #here name is each element in the category
            c+=1
    #print(f"In the {name} category, {c} apps are there")
    categories[name] = c
print(categories)

print("--------------------------------")

#The number of apps in Type
print(df['Type'].unique())   
Types = {}
for name in df['Type'].unique():  #this code allows looping unique values(Free and Paid)
    t = 0
    for i in df['Type']:          #looping starts from Free and then Paid
        if(i==name):              #Here the name is each element in the Type(i.e., Free and Paid)
            t+=1
    Types[name]=t
print(Types)

print("----------------------------------")
#The number of apps in Content Rating
print(df['Content Rating'].unique())
CR = {}
for name in df['Content Rating'].unique():
    p = 0
    for i in df['Content Rating']:
        if(i==name):
            p+=1
        CR[i]=p
print(CR)
print("----------------------------")
df['Category'].describe()
```

```python
df.fillna(method='pad',inplace=True)  #this temporarily fills the value and gives the complete overview that the data looks alike
#the method "pad" replaces the missing values with the value above to it 
#the method "bfill" replaces the missing values with the values below to it.
#inplace=true will permanently change the missing values to the values we gave.
```

```python
#Handling errors in Categorical Data, here we select all the columns and add most frequent values to the missing values
#but in Numerical type data we have to select particular column and there we replace the missing value by most_frequent value(mode) or mean.. 
imputer = SimpleImputer(missing_values=np.nan, strategy='most_frequent')
imputer.fit(df.iloc[:,:].values)
df.iloc[:,:] = imputer.transform(df.iloc[:,:].values)
df

#pros: No data loss, mantains Dataset Structure
#mode - is good for small datasets with repeating categories.
#Unknown - useful when missing values need to be explicitly labeled.
#Custom placeholders : Can be domain-specific.
```
